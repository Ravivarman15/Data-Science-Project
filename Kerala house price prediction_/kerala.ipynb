```python
final_features = ['District',
                  'Property_Type',
                  'BHK',
                  'Builtup_Area_sqft',
                   'Land_Area_cent', 
                   'Age_of_Property',
                   'Bathrooms', 
                   'Parking_Slots',
                   'Distance_to_City_km']
target = df['House_Price_Lakhs']
```

```python
df.Property_Type.unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df[final_features],target,train_size=0.8,random_state=42)
```

```python
cat_cols = ['District','Property_Type']
num_cols = ['BHK','Builtup_Area_sqft',
                   'Land_Area_cent', 
                   'Age_of_Property',
                   'Bathrooms', 
                   'Parking_Slots',
                   'Distance_to_City_km']
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('cat',OneHotEncoder(drop='first',handle_unknown='ignore'),cat_cols),
        ('num',StandardScaler(),num_cols)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
log_p = np.log1p(df['House_Price_Lakhs'])
```

```python
# df[log_p].fillna(df[log_p].median(),inplace=True)
# df[log_p].isna().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df[final_features],log_p,train_size=0.8,random_state=42)
```

```python
lr_model = Pipeline(steps=[
    ('preprocessing',preprocessor),
    ('model',LinearRegression())
    # ('model',Ridge(alpha=1))

])
rf_model = Pipeline(steps=[
    ('preprocessing',preprocessor),
    ('model',RandomForestRegressor(
        n_estimators=500,
        max_depth=20,
        min_samples_split=10,
        min_samples_leaf=5,
        n_jobs=-1,
        random_state=42
    ))
])
```

```python
lr_model.fit(x_train,y_train)
rf_model.fit(x_train,y_train)
```

```python
lr_pred = lr_model.predict(x_test)
rf_pred = rf_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
import numpy as np
```

```python
def evaluate_model(y_true,y_pred):
    mse = mean_squared_error(y_true,y_pred)
    rmse = np.sqrt(mean_squared_error(y_true,y_pred))
    r2 = r2_score(y_true,y_pred)
    return mse,rmse,r2
```

```python
mse1,rmse1,r21 = evaluate_model(y_test,lr_pred)
mse2,rmse2,r22 = evaluate_model(y_test,rf_pred)
```

```python
eva_df = pd.DataFrame({
    'Model':['LinearRegression','RandomForestRegressor'],
    'MSE':[mse1,mse2],
    'RMSE':[rmse1,rmse2],
    'R2':[r21,r22]
})
```
